```python
def dem(category, buyers, products, orders):
    buyers.rename(columns = {'name':'name_buyer'}, inplace = True)
    buyers.rename(columns = {'id':'id_buyer'}, inplace = True)
    SUMMARY = pd.concat([orders, buyers[["name_buyer", "email", "telnumber", "address"]]], axis=1, join="inner")
    category.rename(columns = {'id':'id_category'}, inplace = True)
    Cat_Prod = pd.concat([category[["category"]], products], axis=1, join="inner")
    SUMMARY = pd.concat([SUMMARY, Cat_Prod[["category", "id_category", "name", "price", "count"]]], axis=1, join="inner")
    
    return SUMMARY
```

```python
def cat_price(dem):
    uniq_cat = dem["category"].unique()
    df = {"category":[], "max_price": []}
    for cat in uniq_cat:
        tmp = dem[dem.category == cat][["category", "price"]].nlargest(1, "price")
        df["category"].append(cat)
        df["max_price"].append(tmp["price"].iloc[0])
    res = pd.DataFrame(df)
    return res
        
```
